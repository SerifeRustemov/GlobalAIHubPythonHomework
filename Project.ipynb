Simple Student Management System

```python
# Lessons are taken from the user

def set_course_names():
    print("Please enter 5 courses names:")
    i = 0
    courses = []
    while i < 5:

        course = input("{}. course name: ".format(i+1)).capitalize()

        if course not in courses and course is not None:
            courses.append(course)
            i += 1
            
        else:
            print(course , "is already in course list.")
            print("Please enter a different course")

    return courses
```

```python
#transfer grade to letter sign

def course_grade_result(total):

    if (total >= 90):
        return "AA"
    elif (70 <= total) and (total < 90):
        return "BB"
    elif (50 <= total) and (total < 70):
        return "CC"
    elif 30 <= total and total < 50:
        return "DD"
    else:
        return "FF"
```

```python
def display_courses(courses):
    lenght = len(courses) 

    for i in range(lenght):
        c = courses[i]
        print("{}- {}".format(i+1 , c))
```

```python
#enter user name and surname
#User is asked whether his/her name and surname are correct.
#User can enter his/her name and surname only 3 times.
# displayed welcome message

def enter_name_surname():
    count = 3
    while True:
        student_name = input("Student name:").capitalize()
        student_surname = input("Student surname:").capitalize()

        correct = input("Are your name and surname correct? Please enter yes or no: ").upper()
        print()


        if correct == 'YES':
            print()
            print("****************************")
            print("Welcome {} {}!".format(student_name,student_surname))
            print("****************************")
            break

        elif correct == 'NO':
            count -= 1

            if count == 0:
                return "Please try again later"
                

            print("Please enter student name and surname again!")

        else:
            print("Invalid input! You have to enter 'yes' or 'no' after your name and surname")
            print()

    return student_name, student_surname
```

```python
#Student class is created.
#It holds student name,surname,courses that student took
#Calculated one course's grade which student selected


class Student():
    
    def __init__(self, student_name, student_surname):
        self.student_name = student_name
        self.student_surname = student_surname
        self.student_courses = []
        self.grades = {}
    
    def select_courses(self,courses):
        
        i = 0
        x = 'Y'
        while i < 5 and x == 'Y':
            s_course = input("{}. course: ".format(i+1)).capitalize()

            if (s_course in courses) and (s_course not in self.student_courses) :
                self.student_courses.append(s_course)
                i += 1

            else:
                print("This course is already in your courses or invalid input. Please enter course name again!!!")

            
            if i < 5:
                while True:
                    x = input("Do you want to add another course? (Y/N)").upper()
                    if x not in {'Y','N'}:
                        print("Invalid input!")
                    else:
                        break
                    
        return self.student_courses
        
           
    def set_grades(self, course_name):
        while True:
            try:
                midterm = int(input("Midterm grade: "))
                self.grades['midterm'] = midterm
                break
            except ValueError:
                print("Invalid input!")
                
        while True:
            try:
                final = int(input("Final grade: "))
                self.grades['final'] = final
                break
            except ValueError:
                print("Invalid input!")
        
        while True:
            try:
                project = int(input("Project grade: "))
                self.grades['project'] = project
                break
            except ValueError:
                print("Invalid input!")
        
    
    def course_total_grade(self):
        m = self.grades['midterm']
        f = self.grades['final']
        p = self.grades['project']
        
        total = (m * 0.3) + (f * 0.5) + (p * 0.2)
        return total
```

```python
course_list = set_course_names()

name, surname = enter_name_surname()

student = Student(name,surname)
              
display_courses(course_list)

print()
print("Please enter courses which you take from the list above:")

stu_courses = student.select_courses(course_list)

print()
if len(stu_courses) <3:
    print("You failed in class! Because you have to take minimum 3 courses.")
    
else:
    while True:
        s_course = input("Enter your course which you want to calculate its grade:").capitalize()
        
        if s_course not in stu_courses:
            print('Invalid input!')
        else:
            break
        
    student.set_grades(s_course)
    t = student.course_total_grade()
    grade = course_grade_result(t)
    
    print()
    if grade == 'FF':
        print('Your {} course grade is FF and you failed.'.format(s_course))
    else:
        print("Congrats!! Your {} course grade is {} and you passed this course.".format(s_course, grade))
```
